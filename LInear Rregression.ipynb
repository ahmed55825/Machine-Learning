```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Sample data
X = np.array([1416, 1600, 2104, 2400, 3000])  # Input feature
Y = np.array([232, 330, 400, 369, 540])  # Output
```

```python
# Split the data into training and testing sets
split_ratio = 0.8  # 80% for training, 20% for testing
split_index = int(len(X) * split_ratio)
```

```python
X_train, X_test = X[:split_index], X[split_index:]
Y_train, Y_test = Y[:split_index], Y[split_index:]
```

```python
# Calculate the necessary statistics for training data
n = len(X_train)
mean_x = np.mean(X_train)
mean_y = np.mean(Y_train)
```

```python
# Calculate the slope (w) and intercept (b) of the regression line
numer = np.sum((X_train - mean_x) * (Y_train - mean_y))
denom = np.sum((X_train - mean_x) ** 2)
w = numer / denom
b = mean_y - w * mean_x
```

```python
# Create the regression equation: Y = wX + b
```

```python
# Predict for the test set
Y_pred = w * X_test + b
```

```python
# Visualization
plt.scatter(X_train, Y_train, color='blue', label='Training Data')
plt.scatter(X_test, Y_test, color='green', label='Testing Data')
plt.plot(X_train, w * X_train + b, color='red', label='Regression Line')



plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression')
plt.legend()
plt.grid(True)
plt.show()
```

```python
print(f"Slope (w): {w}")
```

```python
print(f"Intercept (b): {b}")
```

```python
print(f"Regression Equation: Y = {w:.2f}X + {b:.2f}")
```

```python
# Evaluate the model (e.g., using Mean Squared Error)
mse = np.mean((Y_pred - Y_test) ** 2)
print(f"Mean Squared Error on Test Data: {mse:.2f}")
```
