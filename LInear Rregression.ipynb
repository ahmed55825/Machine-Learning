```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Sample data
X = np.array([1416, 1600, 1801, 2020, 2104, 2355, 2400, 2650, 2800, 2964, 3000])  # Input feature
Y = np.array([232, 250, 295, 330, 365, 400, 369, 351, 214,500, 540])  # Output
```

```python
# Hyperparameters
learning_rate = 0.0001  # Reduce learning rate
num_iterations = 10000  # Increase the number of iterations
```

```python
# Standardize the input features to improve stability
X_mean = np.mean(X)
X_std = np.std(X)
X_scaled = (X - X_mean) / X_std
```

```python
# Initialize model parameters with reasonable values
w = 0.5
b = 100
```

```python
# Number of data points
n = len(X)
```

```python
# Gradient Descent
for iteration in range(num_iterations):
    # Predicted values using the current model parameters
    Y_pred = w * X_scaled + b
    
    # Compute gradients with respect to w and b
    gradient_w = (-2/n) * np.sum(X_scaled * (Y - Y_pred))
    gradient_b = (-2/n) * np.sum(Y - Y_pred)
    
    # Update model parameters using gradients and learning rate
    w = w - learning_rate * gradient_w
    b = b - learning_rate * gradient_b
```

```python
# Create the regression equation: Y = wX + b

# Predict a new value
new_x = 3500
```

```python
# Scale the new_x value using the same mean and std
new_x_scaled = (new_x - X_mean) / X_std
predicted_y = w * new_x_scaled + b
```

```python
# Visualization
plt.scatter(X, Y, color='blue', label='Data Points')
plt.plot(X, w * X_scaled + b, color='red', label='Regression Line')
plt.scatter(new_x, predicted_y, color='green', marker='x', s=100, label='Predicted Point')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression (Gradient Descent)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
print(f"Weight (w): {w:.2f}")
print(f"Intercept (b): {b:.2f}")
print(f"Regression Equation: Y = {w:.2f}X + {b:.2f}")
print(f"Predicted Y for X = {new_x}: {predicted_y:.2f}")
```
